```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline
```

```python
# Load history
history = np.load('models/PIR_CNN_Model3_6_19052021211410_history.npy', allow_pickle=True).item()

# summarize history for accuracy
plt.plot(history['sparse_categorical_accuracy'])
plt.plot(history['val_sparse_categorical_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
print(np.argmax(history['val_sparse_categorical_accuracy']))
print(np.max(history['val_sparse_categorical_accuracy']))
print(history['val_loss'][np.argmax(history['val_sparse_categorical_accuracy'])])
```
